```python
from PIL import Image
import os

def get_unique_image_sizes(directory):
    sizes = set()
    for filename in os.listdir(directory):
        if filename.lower().endswith(('.png', '.jpg')):
            try:
                with Image.open(os.path.join(directory, filename)) as img:
                    sizes.add(img.size)  # img.size is a tuple (width, height)
            except IOError:
                print(f"Could not process file {filename}")
    return sizes

# Replace 'your_image_directory' with the path to your image dataset
image_sizes = get_unique_image_sizes('/home/user/WindowsShare/05. Data/02. Training&Test/013. General Concrete Damage/01. Cityscapes/v_bh/leftImg8bit/train')
print(image_sizes)
```
